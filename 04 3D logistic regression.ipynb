<h4>Load lib</h4>

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from time import time
from collections import Counter
from sklearn import datasets
from scipy.stats import norm
from scipy.integrate import simpson
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, mean_squared_error
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.linear_model import LinearRegression, LogisticRegression
```

```python
RANDOM_STATE = 24
np.random.seed(RANDOM_STATE)

NOISE = 0.1
N_SAMPLES =1000
```

```python
params = {
    'legend.fontsize': 'small',
    'figure.figsize': (15,5),
    'axes.labelsize': 'medium',
    'axes.titlesize': 'large',
    'xtick.labelsize': 'medium',
    'ytick.labelsize': 'medium',
    'savefig.dpi': 75,
    'image.interpolation': 'none',
    'savefig.bbox': 'tight',
    'lines.linewidth': 1,
    'legend.numpoints': 1,
}

CMAP = plt.cm.rainbow
plt.rcParams.update(params)
plt.set_cmap(CMAP)
```

<h4>Load dataset</h4>

```python
X,y=datasets.load_iris(return_X_y=True)
label_names={0:'setosa',1:'versicolor',2:'virginica'}
X.shape,y.shape
```

<h4>One vs Rest</h4>

```python
from sklearn.multiclass import OneVsRestClassifier
clf = OneVsRestClassifier(LogisticRegression(penalty = None,
                        random_state=RANDOM_STATE))
clf.fit(X, y)

y_pred = clf.predict(X)

x_err = X[:, 0][y != y_pred]
y_err = X[:, 1][y != y_pred]
x_err.shape, y_err.shape
```

```python
#plotting
fig,ax = plt.subplots(figsize=(15, 12))
title = 'Iris Dataset'
fig.suptitle(title)

ax.scatter(X[:, 0], X[:,1], s = 40, c= y , cmap=CMAP)
ax.scatter(x_err, y_err,
          color = 'w',
          s = 100,
          alpha=0.8,
          edgecolors='k'
          )
```

<h4>Multinomial</h4>

```python
clf = LogisticRegression(penalty = None,
                        random_state=RANDOM_STATE)
clf.fit(X, y)

y_pred = clf.predict(X)

x_err = X[:, 0][y != y_pred]
y_err = X[:, 1][y != y_pred]
x_err.shape, y_err.shape
```

```python
fig,ax = plt.subplots(figsize=(15, 12))
title = 'Iris Multinomial Dataset'
fig.suptitle(title)

ax.scatter(X[:, 0], X[:,1], s = 40, c= y , cmap=CMAP)
ax.scatter(x_err, y_err,
          color = 'w',
          s = 100,
          alpha=0.8,
          edgecolors='k'
          )
```
